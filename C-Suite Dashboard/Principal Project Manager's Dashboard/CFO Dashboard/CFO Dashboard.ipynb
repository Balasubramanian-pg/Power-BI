```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Function to generate a list of dates (weekly)
def generate_dates(start_date, weeks=104):
    dates = [start_date + timedelta(weeks=week) for week in range(weeks)]
    return pd.to_datetime(dates)

# Function to generate random financial data based on logical constraints
def generate_financial_data(weeks=104):
    # Date generation
    start_date = datetime(2022, 4, 1)  # Start date for 2 years
    dates = generate_dates(start_date, weeks)
    
    # Sample data generation for different metrics
    total_revenue = np.random.randint(100000, 200000, size=weeks)
    net_profit = np.random.randint(20000, 50000, size=weeks)
    operating_cash_flow = np.random.randint(10000, 30000, size=weeks)
    gross_margin = np.random.uniform(30, 50, size=weeks)
    
    ebitda = np.random.randint(30000, 60000, size=weeks)
    operating_income = np.random.randint(20000, 50000, size=weeks)
    cogs = np.random.randint(30000, 80000, size=weeks)
    
    operating_cash_flow_2 = np.random.randint(10000, 30000, size=weeks)
    free_cash_flow = np.random.randint(5000, 15000, size=weeks)
    financing_cash_flow = np.random.randint(-10000, 10000, size=weeks)
    investing_cash_flow = np.random.randint(-5000, 5000, size=weeks)
    
    debt_to_equity = np.random.uniform(0.2, 2.0, size=weeks)
    return_on_assets = np.random.uniform(5, 15, size=weeks)
    return_on_equity = np.random.uniform(8, 20, size=weeks)
    current_ratio = np.random.uniform(1, 3, size=weeks)
    
    total_expenses = np.random.randint(50000, 150000, size=weeks)
    operational_expenses = np.random.randint(30000, 100000, size=weeks)
    capital_expenditures = np.random.randint(10000, 50000, size=weeks)
    top_5_expenses = np.random.randint(1000, 5000, size=(weeks, 5))
    
    revenue_by_product = np.random.randint(10000, 50000, size=(weeks, 3))
    revenue_by_region = np.random.randint(20000, 60000, size=(weeks, 3))
    revenue_growth = np.random.uniform(5, 15, size=weeks)
    
    audit_findings = np.random.randint(0, 5, size=weeks)
    compliance_status = np.random.choice([True, False], size=weeks)
    
    forecasted_revenue = np.random.randint(100000, 200000, size=weeks)
    actual_revenue = forecasted_revenue + np.random.randint(-10000, 10000, size=weeks)
    variance = actual_revenue - forecasted_revenue
    forecast_accuracy = (actual_revenue / forecasted_revenue) * 100
    
    tax_liabilities = np.random.randint(10000, 50000, size=weeks)
    current_liabilities = np.random.randint(10000, 30000, size=weeks)
    long_term_liabilities = np.random.randint(50000, 100000, size=weeks)
    
    industry_benchmark_roa = np.random.uniform(5, 15, size=weeks)
    company_performance_roa = np.random.uniform(5, 20, size=weeks)
    performance_vs_industry = (company_performance_roa / industry_benchmark_roa) * 100
    
    # Consolidating data into DataFrame
    df = pd.DataFrame({
        'Date': dates,
        'Total Revenue': total_revenue,
        'Net Profit': net_profit,
        'Operating Cash Flow': operating_cash_flow,
        'Gross Margin': gross_margin,
        'EBITDA': ebitda,
        'Operating Income': operating_income,
        'COGS': cogs,
        'Operating Cash Flow (2)': operating_cash_flow_2,
        'Free Cash Flow': free_cash_flow,
        'Cash Flow from Financing': financing_cash_flow,
        'Cash Flow from Investing': investing_cash_flow,
        'Debt to Equity Ratio': debt_to_equity,
        'Return on Assets (ROA)': return_on_assets,
        'Return on Equity (ROE)': return_on_equity,
        'Current Ratio': current_ratio,
        'Total Expenses': total_expenses,
        'Operational Expenses (OPEX)': operational_expenses,
        'Capital Expenditures (CAPEX)': capital_expenditures,
        'Top 5 Expense Categories': top_5_expenses.tolist(),
        'Revenue by Product Line': revenue_by_product.tolist(),
        'Revenue by Region': revenue_by_region.tolist(),
        'Revenue Growth': revenue_growth,
        'Audit Findings': audit_findings,
        'Compliance Status': compliance_status,
        'Forecasted Revenue': forecasted_revenue,
        'Actual Revenue': actual_revenue,
        'Variance': variance,
        'Forecast Accuracy': forecast_accuracy,
        'Tax Liabilities': tax_liabilities,
        'Current Liabilities': current_liabilities,
        'Long-Term Liabilities': long_term_liabilities,
        'Industry Benchmark ROA': industry_benchmark_roa,
        'Company Performance ROA': company_performance_roa,
        'Performance vs Industry': performance_vs_industry
    })
    
    return df

# Save the DataFrame to the specified path
df = generate_financial_data()
df.to_excel(r"F:\Flipcarbon\11. November\C-Suite Dashboards\financial_dashboard_data.xlsx", index=False)
```
